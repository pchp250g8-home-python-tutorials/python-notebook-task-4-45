Коллекция марок состоит из трёх альбомов. В первом из альбомов находится одна пятая всех марок, 
во втором - несколько седьмых и в третьем - 303 марки. Сколько марок в коллекции.

```python

from sympy import Eq
from sympy import solve
from sympy import symbols

x, y = symbols("x y")
eq1 = Eq(1 / 5 * x + x * y / 7 + 303, x)
eq2 = Eq(y, 5)
sol = solve([eq1, eq2], [x, y])
print(sol)
```
